```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**PIE CHART**

```python
flavors = ('Chocolate', 'Vanilla', 'Pistachio')
votes = (11, 13, 15)
colors = ('#A52A2', '#FFEBCD', '#000000')
explode = (0,0,0)

plt.title('Favorite Ice Cream Flavors')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True
)
plt.show()
```

**bar chart**

```python
import pandas as pd

countries = ['USA', 'China', 'India', 'Japan']
populations = [331002651, 1444216107, 1380004385, 126476461]

df = pd.DataFrame({
    'Country': countries,
    'Population': populations,
})

df.sort_values(by='Population', inplace=True)

x_cords = np.arange(len(df))

plt.bar(x_cords, df['Population'], tick_label=df['Country'])
plt.ylabel('Population')
plt.title('Asean Populations')

plt.show()
```

**LINE GRAPH**

```python
import matplotlib.pyplot as plt

hour = [0, 1, 1, 2, 0, 3]
temperature_c = [0, 1, 2, 0, 3, 5]

plt.plot(hour, temperature_c)
plt.xlabel('Hour')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Variation over Time')
plt.grid(True)
plt.show()
```
